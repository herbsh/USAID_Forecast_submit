```python
import pandas as pd 
contraceptive_logistics=pd.read_csv('../0_data/contraceptive_logistics_data.csv') 
contraceptive_logistics.head() 
```

```python
contraceptive_logistics.describe()
```

```python
contraceptive_logistics['day']=1
contraceptive_logistics['date']=pd.to_datetime(contraceptive_logistics[['year','month','day']])
contraceptive_logistics=contraceptive_logistics.sort_values('date')
contraceptive_logistics.set_index('date',inplace=True )
contraceptive_logistics.head(2)
```

```python
df=contraceptive_logistics.copy(deep=True)
```

```python
df_resample=df.groupby(['site_code','product_code']).resample('MS').asfreq() 
               
```

```python
df_resample=df_resample.drop(['product_code','site_code'],1).reset_index()
```

```python
df_resample.head()
```

```python
df_resample.tail()
```

```python
df_resample.describe()
```

```python
import datetime
df_resample['date']=pd.to_datetime(df_resample['date'])
df_resample['year'], df_resample['month']=df_resample['date'].apply(lambda x:x.year),df_resample['date'].apply(lambda x:x.month)
```

```python
df_resample.head()
```

```python
col_to_export=['site_code','product_code','date','year','month','region','district','stock_distributed']
df_resample=df_resample[col_to_export]
```

```python
df_resample.head()
```

```python
df_resample['stock_distributed']=df_resample['stock_distributed'].fillna(0)
```

```python
df_resample.head(30)
```

### Fillna , use merge just to be safe 

```python
site_info=df_resample[['site_code','region','district']].drop_duplicates().dropna()
```

```python
site_info.head()
```

```python
df_resample=pd.merge(left=df_resample.drop(columns=['region','district']),right=site_info,on='site_code',how='left')
```
